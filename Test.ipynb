```python
import requests
import base64
from PIL import Image
from io import BytesIO

url_api = "https://tarea-2-4ivb.onrender.com/predict"

# Imagen por URL
payload = {
    "image_url": "https://www.gastronomiavasca.net/uploads/image/file/3268/salmon.jpg"
}

resp = requests.post(url_api, json=payload)
data = resp.json()

print("Label:", data["label"])
print("Score:", data["score"])
print("Decision:", data["decision"])
print("Probabilidades:", data["probs"])

# Mostrar miniatura
if data["image_thumb_base64"]:
    img_bytes = base64.b64decode(data["image_thumb_base64"])
    img = Image.open(BytesIO(img_bytes))
    display(img)

```
